```python
from diffusers import DiffusionPipeline
```

```python
generator = DiffusionPipeline.from_pretrained('CompVis/ldm-text2im-large-256')
```

```python
generator.to('cuda')
```

```python
prompt = "An image of dog in Picaso style"
result = generator([prompt], num_inference_steps=50, eta=0.3, guidance_scale=6)
```
